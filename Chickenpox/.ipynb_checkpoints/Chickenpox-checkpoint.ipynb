## About the dataset:

<br>


### Sample Design:

The **NIS-Child** uses two phases of data collection to obtain vaccination information for a large national probability sample of young children: an **RDD telephone survey** designed to identify households with children 19 through 35 months of age, followed by the **Provider Record Check, a mailed survey to childrenâ€™s vaccination providers.**

<br>


### Data Collection:

A total of **35,703 household interviews** comprised the **National Immunization Survey (NIS)** data from 2022 (latest available).  

The total cellular phone sample consisted of 25,057,932 telephone numbers. Of those, 38,512 were eliminated before release to the telephone centers as numbers on the NIS do-not-call list. The remaining 25,019,420 numbers were sent to the telephone centers to be dialed, and 1,717,879 active personal cellular phone numbers (APCNs) were identified, as shown in Row F. Among the identified APCNs, 1,457,051 (84.8%) were successfully screened. Of these, 49,111 (3.4%) were deemed eligible for the NIS-Child interview. Among the identified eligible respondents, 33,450 (68.1%) completed the interview. 

<br>


### Objective of this study:

Identify **factors that are associated with the number of Varicella doses** using **kruskal-Wallis** and **Multinomial Logistic Regression**

# Kruskal-Wallis

Let's use Kruskal-Wallis due to violations of ANOVA assumptions

"The Kruskal Wallis test and other non-parametric (or distribution-free) tests are useful to test hypotheses when the assumption for normality of the data does not hold"

```python
# Define a function for the summary
def summary_cont(var):
    tb = rp.summary_cont(data_subset_2['P_NUMVRC'].groupby(data_subset_2[var]))
    return tb
```

```python
# define a function to perform the Kruskal-Wallis H-test

def kruskal_test(var):     
    
    H_statistic, kruskal_p_value = stats.kruskal(*[data_subset_2['P_NUMVRC'][data_subset_2[var] == x]
                                               for x in data_subset_2[var].unique()])
    significance_level = .05
    result = 'No' if kruskal_p_value > significance_level else ''
    
    summary = f'The p-value of {kruskal_p_value} sugests that there is {result} statistically \
significance differences in the median of doses for different {var} groups'
 
   # print(f'H-Statistic: {H_statistic}\np_value: {kruskal_p_value},\nResult: {result}')
    print(summary)
```

 **$ H_0$**: The distributions of the number of doses for all age groups are identical.
 
 **$ H_1$**: At least one age group has a different distribution of number of doses.

# Multinomial Logistic Regression

### Create series of dummy dichotomous variables

```python
vars = ['SEX','AGEGRP', 'RACE_K', 'INCPOV1', 'EDUC1', 'INCPORAR', 'CEN_REG']

data_enc = pd.get_dummies(data_subset[vars], dtype=int )
```

```python
data_enc['P_NUMVRC'] = data_subset['P_NUMVRC']
```

```python
# remove missing values for number of doses
data_enc_2 = data_enc[data_enc['P_NUMVRC'].notna()]
```

```python
# Remove characters from column names
data_enc_2.columns = [re.sub('[- + ,$]','',col).replace('>', 'greater').replace('<=', 'samller') for col in data_enc.columns]  
```

```python
plt.figure(figsize=(30, 20))
sns.heatmap(pd.DataFrame(
                data_enc_2, 
                columns=data_enc_2.columns).corr(), 
                annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Due to multicolinearity let's check the first category for each variable

```python
pd.set_option('display.max_rows', None)

warnings.filterwarnings('ignore')

glm_formula = 'P_NUMVRC ~ C(SEX_Male) + C(AGEGRP_1923Months) + C(RACE_K_BlackOnly) + C(INCPOV1_AbovePovertysamller75K) + C(EDUC1_12Years) + C(CEN_REG_Midwest)'
 

model = smf.mnlogit(glm_formula, data=data_enc_2)
res = model.fit()
print(res.summary())
```
